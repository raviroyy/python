```python
#function with no argument and no return type
l=int(input("enter l= "))   #global variable
b=int(input("enter b= ")) 
def cal():
    a=l*b
    print(a)
cal()
print(l,b)
```

```python
#function with no argument and no return type
def cal():
    global l,b,a
    l=int(input("enter l= "))   #local variable because defined inside function so cant be used outside function
    b=int(input("enter b= "))   #local variable
    a=l*b
    print(a)
cal()
print(l,b)
```

```python
#function with argument and no return type
def cal(l,b):#argument
    a=l*b
    print(a)
    pass
l=int(input("enter l= "))
b=int(input("enter b= ")) 
cal(l,b)
    
```

```python
def cal(x,y):#parameter
    a=x*y
    print(a)
    pass
l=int(input("enter l= "))
b=int(input("enter b= ")) 
cal(l,b)  #argument
```

```python
def cal(data):#parameter
    x,y,z=data
    v=x*y*z
    print(v)
    pass
l=int(input("enter l= "))
b=int(input("enter b= ")) 
h=int(input("enter h= "))
data=(l,b,h)
cal(data)  #argument
```

```python
#function with no argument and return type
def cal():
    l=10
    b=5
    a=l*b
    return a
print(cal())
```

```python
#function with no argument and return type
def cal():
    l=10
    b=5
    a=l*b
    return a
h=2
area=cal()
volume=area*h
print(volume)
print(area)
```

```python
#function with no argument and return type
def cal():
    l=10
    b=5
    h=2
    a=l*b
    v=a*h
    return a,v
x=cal()
ar,vo=x
print("area= ",ar)
print("volume= ",vo)
```

```python
#function with argument and return type
def cal(l,b,h):
    a=l*b
    v=a*h
    return a,v
l=10
b=5
h=2
    
x=cal(l,b,h)
ar,vo=x
print("area= ",ar)
print("volume= ",vo)
```

```python
def cal():
    a=int(input("enter the value of a= "))
    b=int(input("enter the value of b= "))
    o=input("enter + - * /")
    if o== "+":
        print(a+b)
    elif o=="-":
        print(a-b)
    elif o=="*":
        print(a*b)
    elif o=="/":
        print(a/b)
    else:
        print("invalid operator")
    
    x=input("enter y for more cal")
    if x=='y':
        cal()
cal()
            
            
```

```python

    
def add(a,b):
    return(a+b)
def sub(a,b):
    return(a-b)
def mul(a,b):
    return(a*b)
def div(a,b):
    return(a/b)
def cal():
    a=int(input("enter the value of a= "))
    b=int(input("enter the value of b= "))
    o=input("enter + - * /")
    if o== "+":
        print(add(a,b))
    elif o=="-":
        print(sub(a-b))
    elif o=="*":
        print(mul(a,b))
    elif o=="/":
        print(div(a,b))
    else:
        print("invalid operator")
    
    x=input("enter y for more cal")
    if x=='y':
        cal()
cal()
```
